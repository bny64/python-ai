```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)
```

```python
# 음성 파일 업로드 및 텍스트 변환 (실제 테스트할 음성 파일 경로를 입력하세요)
audio_file_path = "./audio/lsy_audio_2023_58s.mp3"

if os.path.exists(audio_file_path):
    print("음성 파일을 업로드하는 중...")
    audio_file = client.files.upload(file=audio_file_path)
    
    print("텍스트로 변환하는 중...")
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[audio_file, "이 오디오 파일의 내용을 텍스트로 변환해줘."]
    )
    
    print("\n[변환 결과]")
    print(response.text)
else:
    print(f"'{audio_file_path}' 파일이 존재하지 않습니다. 실제 음성 파일 경로를 설정해주세요.")
```
