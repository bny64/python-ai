```python
from google import genai
from dotenv import load_dotenv
from google.genai import types

import os
import io
import urllib.request
from PIL import Image

load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")

client = genai.Client(api_key=api_key)
```

### 방법 1: PIL(Pillow) 이미지 객체 사용하기
가장 보편적인 방법으로, URL에서 이미지를 다운로드하여 `PIL.Image` 객체로 변환한 후 `contents`에 직접 전달합니다.

```python
from google.genai import types

image_url = "https://i.namu.wiki/i/Y7hSqgJ0zdI8EoSUs9sYyB7XgPSXvBX8YhmW4i0URSpWAzczd3swIl0WuRVScpHCUxFvNIRUP_9pdiSG8lOxLQ.webp"

response = client.models.generate_content(
    model="gemini-3.1-flash-lite",
    contents=[
        types.Part.from_uri(
            file_uri=image_url,
            mime_type="image/jpeg"  # 이미지의 MIME 타입을 지정해야 합니다.
        ),
        "이 이미지에 무엇이 보이나요?"
    ]
)
print(response.text)
```

```python
import base64


def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


image_path = "../data/background.jpg"

# 이미지를 base64로 인코딩
base64_image = encode_image(image_path)

print(base64_image)
```

```python
import base64
from google.genai import types

# base64 문자열을 바이트로 디코딩하여 Gemini API에 전달
response = client.models.generate_content(
    model="gemini-3.1-flash-lite",  # 혹은 사용 중이신 모델 (예: gemini-2.5-flash 등)
    contents=[
        types.Part.from_bytes(
            data=base64.b64decode(base64_image),
            mime_type="image/jpeg"  # 배경 이미지 파일 형식(jpg)에 맞게 설정
        ),
        "이 이미지에 무엇이 보이나요?"
    ]
)

print(response.text)
```
